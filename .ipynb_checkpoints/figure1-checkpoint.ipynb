```python
#extracting columnns of interest from d
d = d[['word', 'freq', 'ivc_avg', 'Entropy_Roark', 'Surprisal_Roark',
       'env_avg','is_key','xmax', 'xmin', 'entropy']]
#adding average time
d['xavg'] = (d['xmax']+d['xmin'])/2
```

```python
#set s, can be any sentence
s = d.iloc[614:624].reset_index()


s['xavg0'] = s['xavg'] - s['xmin'].min() #correct time
s['freq'] = s['freq'].replace([0.0],0.2)
s['xmin0'] = s['xmin'] - s['xmin'].min() #correct time
s['xmax0'] = s['xmax'] - s['xmin'].min() #correct time

s
```

```python
#getting ivc up to some index
#m = ivc_env.loc[5390:8849].reset_index()
m = ivc_env.loc[s['xmin'].min() * 100 : s['xmax'].max()*100].reset_index()
#m['time'] = m['time']/10 - s['xavg'].min()*10 #correcting time
m['time'] = m['time'] - s['xmin'].min() #correcting time to start at 0


#rolling average
window_size = 5
m['rolling_avg'] = m['ivc_standarized'].rolling(window_size).mean()
#replacing window-size nans at start
m['rolling_avg'] =  m['ivc_standarized'][:window_size-1].append( m['rolling_avg'][window_size-1:])

m
```

```python
#getting audio env up to some index
#a = aud_env.loc[5390:8849].reset_index()
a = aud_env.loc[s['xmin'].min() * 100 : s['xmax'].max()*100].reset_index()
#a['time'] = a['time'] - s['xavg'].min()*10 #correcting time

a['time'] = a['time'] - s['xmin'].min() #correcting time to start at 0


#rolling average
window_size = 5
a['rolling_avg'] = a['envelope_standardized'].rolling(window_size).mean()
#replacing window-size nans at start
a['rolling_avg'] =  a['envelope_standardized'][:window_size-1].append( a['rolling_avg'][window_size-1:])

a
```

## Extracting and Cropping Frames of interest

Frames corresponding to the mean timing of a word get extracted from the video, cropped according to cropping parameters and saved into the outputDirectory 

```python
#cropping parameters
y = 75 #100
x = 260 #240 #215
h = 500 # 400
w = 230 #250 #300

#get frames of interest in msec

#middle of words
#frametimes = np.round(np.array(s.xmax + s.xmin) / 2, 2) * 1000 

#even spacing in time
n = 13 # 12 # 10
frametimes = np.arange(s.xmin.min(), s.xmax.max(), s.xmax0.max()/n) * 1000


#set directories
workingDirectory = 'C:/Users/gffun/OneDrive/Desktop/spyder-py3XF/fMRI Intern'
outputDirectory = workingDirectory + '/framesCropped2'

os.chdir(workingDirectory)

#initialize capture object with cv2
#make sure to include the appropriate file 01-16
cap = cv2.VideoCapture('Stimuli_Samir_actual_MPEG/04_Samir_blieb.mpeg')

#initialize output directory, if not already existing
if not os.path.exists(outputDirectory):
    os.mkdir(outputDirectory)
os.chdir(outputDirectory)

#clear output directory
for f in os.listdir(outputDirectory):
    os.remove(os.path.join(outputDirectory, f))

#get, crop and save frames
for i in range(len(frametimes)):
    cap.set(0, frametimes[i]) #set msec of video here
    ret, frame = cap.read() #get frame
    crop_img = frame[y:y+h, x:x+w] #crop frame
    if i < 10:    
        cv2.imwrite('frame0'+str(i)+'.jpeg', crop_img) #save frame
    else:
        cv2.imwrite('frame'+str(i)+'.jpeg', crop_img) #save frame


"""
while(cap.isOpened()):
    ret, frame = cap.read()

    #gray = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    crop_img = frame[y:y+h, x:x+w]
    
    cv2.imshow('frame',crop_img)
    
    if cv2.waitKey(0) & 0xFF == ord('q'): #waitKey 0 waits indefinately for key press to show next pic
        break
"""

#frames = cap.get(5)
#msec = cap.get(0)
#cap.release()
#cv2.destroyAllWindows()
```

## Figure 1
An example sentence from the utilized story, illustrating measures used in GLM analysis. Top: frames cropped from the stimulus video in equidistant intervals. Upper center: word-wise measures of interest entropy and surprisal (in bits) and frequency as a measure of no interest (in arbitrary units). Lower center: z-scored measures of no interest relating to the actor's movement (IVC, left scale) and auditory speech envelope (right scale), both word-wise and as continuous measures (sampling frequency: 100 Hz). For illustrative purposes, a running average with window size 5 for continuous measures was used. Bottom: Distributions of both key- and non-keywords across surprisal and entropy measures.
